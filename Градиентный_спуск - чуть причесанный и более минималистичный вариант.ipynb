```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_name = "/media/art/Terminator/datasets/boston/boston_data.csv"
data = np.genfromtxt(file_name, skip_header=1, delimiter=',')
```

```python
data
```

```python
Y = data[:, -1]
x = data[:, :-1]
```

```python
X = np.zeros(shape=(len(Y), 1 + len(x[0])))
X[:, 0] = 1
```

```python
X[:, 1:] = data[:, :-1]
```

```python
def func_gradient(f, x0, dx=0.0001):
    our_gradient = []
    x = list(x0)
    f1 = f(x)
    for i in range(0, len(x)):
        x[i] += dx
        f2 = f(x)
        x[i] -= dx
        our_gradient.append((f2 - f1) / dx)
        
    return our_gradient
```

```python
def cost(weights):    
    hyp = (X * weights).sum(axis=1)
    return ((hyp - Y)**2).mean()
```

```python
weights = [0] * 14
print(cost(weights))
```

```python
weights = [0] * 14
print("Cost:", cost(weights))
N = 2 * 10**6
lr = 0.0000030

for i in range(N):
    grad = func_gradient(cost, weights)
    weights = [w - lr*gr for w, gr in zip(weights, grad)]
    
    
    if i % 1000 == 0:
        error = cost(weights)
        print("Cost:", error)

```

# Mean Average Error

```python
hyp = (X * weights).sum(axis=1)
print("В среднем предсказанная цена отличается от истинной на", (np.abs(hyp - Y)).mean())
print("Истинная и предсказанная цена")
list(zip(Y, hyp))
```

## Провека с помощью sklearn

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
result = lm.fit(X, Y)
print(result.coef_)
print(result.intercept_)
```

```python
predicted = lm.predict(X)
```

```python
((Y - predicted)**2).sum() / len(Y)
```
